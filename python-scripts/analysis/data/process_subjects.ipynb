## Process the subjects of the experiments and create the train dataset

```python
list_subjects = os.listdir(DATA_DIR_PATH)

print("Found {} subjects to analyze".format(len(list_subjects)))
print("----------------------------------------------------------------------\n")

df_dataset = pd.DataFrame()

for subject in list_subjects:
    if "s" in subject:
        print("Processing subject_id {}".format(subject[1:]))
        df_tmp = process_subject(os.path.join(DATA_DIR_PATH, subject), int(subject[1:]), center_angle_res)
        df_dataset = df_dataset.append(df_tmp)

        print("\n**************************************************************\n")

df_dataset['azimuth'] = df_dataset['azimuth'].astype(int)

# Map the azimuth angles to class labels between 0-180
df_dataset.loc[df_dataset.azimuth < -90, "azimuth"] = -90
df_dataset.loc[df_dataset.azimuth >= 90, "azimuth"] = 89
df_dataset['azimuth'] = df_dataset['azimuth'] + 90
df_dataset.to_csv("{}.csv".format(output_filename), index=False)
```
